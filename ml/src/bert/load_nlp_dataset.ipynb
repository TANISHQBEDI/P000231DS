# BERT Dataset Loader

Starter notebook for loading `NLP_Dataset_2026.xlsx`.

```python
import warnings
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 120)
sns.set_theme(style="whitegrid")
```

```python
dataset_name = "NLP_Dataset_2026.xlsx"
search_roots = [Path.cwd().resolve(), *Path.cwd().resolve().parents]

project_root = next(
    (
        root
        for root in search_roots
        if (root / "ml" / "data" / "raw" / dataset_name).exists()
    ),
    None,
)

if project_root is None:
    raise FileNotFoundError(f"Could not find {dataset_name} from the current notebook location.")

data_path = project_root / "ml" / "data" / "raw" / dataset_name
df = pd.read_excel(data_path)

print(f"Loaded dataset from: {data_path}")
print(f"Shape: {df.shape}")
```

```python
df.head()
```

```python
print(f"Dataset shape: {df.shape}")
print(f"Duplicate rows: {df.duplicated().sum()}")

summary_df = pd.DataFrame({
    "dtype": df.dtypes.astype(str),
    "null_count": df.isnull().sum(),
    "null_percent": (df.isnull().mean() * 100).round(2),
    "unique_count": df.nunique(dropna=True),
})

summary_df = summary_df.sort_values(["null_count", "unique_count"], ascending=[False, False])
summary_df
```

```python
null_counts = summary_df.loc[summary_df["null_count"] > 0, "null_count"].sort_values()

if null_counts.empty:
    print("No null values found in the dataset.")
else:
    plt.figure(figsize=(10, max(6, len(null_counts) * 0.25)))
    plt.barh(null_counts.index, null_counts.values, color="#4C78A8")
    plt.title("Null Counts by Column")
    plt.xlabel("Null count")
    plt.ylabel("Column")
    plt.tight_layout()
    plt.show()
```

```python
null_threshold = 5
null_percent_by_column = summary_df["null_percent"].reindex(df.columns)
columns_to_drop = null_percent_by_column[null_percent_by_column > null_threshold].index.tolist()

df_filtered = df.drop(columns=columns_to_drop).copy()

print(f"Original shape: {df.shape}")
print(f"Filtered shape: {df_filtered.shape}")
print(f"Dropped {len(columns_to_drop)} columns with more than {null_threshold}% null values.")

dropped_columns_df = summary_df.loc[columns_to_drop, ["null_count", "null_percent"]]
dropped_columns_df.sort_values("null_percent", ascending=False)
```
